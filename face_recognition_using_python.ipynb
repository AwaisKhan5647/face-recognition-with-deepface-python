```python
from deepface import DeepFace
import matplotlib.pyplot as plt
import cv2
import json
```

```python
def verify (image1, image2, model_name='VGG-Face'):
  image1=cv2.imread(image1_path) #load images from the path
  image2=cv2.imread(image2_path)
  plt.imshow(image1[:,:,::-1]) # show the loaded image in RGB
  plt.show()
  plt.imshow(image2[:,:,::-1])
  plt.show()
  result = DeepFace.verify(image1,image2,model_name) # DeepFace verify function calling, the ouput will save in the result dictionary
  print('Result =',result)
  verification = result.get("verified") # getting verified key values from the result which can have "True" if face matched else "False"
  if verification:
    print(f"The person is detected as Same, using model {result.get('model')} with confidence of", result.get("confidence"))
  else:
    print(f"The persons are detected as different and not identical, using model {result.get('model')} with confidence of", result.get("confidence"))
```

```python
image1_path = ('/content/Image1.jpeg') # set the path of image 1
image2_path = ('/content/Image2.jpeg') # set the path of image 2
verify(image1_path, image2_path, model_name='Facenet') # model_name = Facenet, VGG-Face, OpenFace, DeepID, ArcFace
```
